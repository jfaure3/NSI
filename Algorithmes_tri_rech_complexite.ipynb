## 2.1. Tri stupide (Stupid sort)

**principe** : on mélange la liste à trier tant qu'elle n'est pas triée.

```python
def est_triée(liste):
    """renvoie si liste est triée ou non
    >>> est_triée([1,2,3])
    True
    >>> est_triée([1,8,3])
    False"""
    for indice in range(len(liste)-1):
        if liste[indice] > liste[indice+1]:
            return False
    return True
doctest.run_docstring_examples(est_triée, globals())        
```

```python
import random
def tri_stupide(liste):
    """renvoie une copie de la liste triée
    >>> tri_stupide([3,2,1])
    [1, 2, 3]
    >>>"""
    liste_tmp=liste[:]
    while not est_triée(liste_tmp):
        random.shuffle(liste_tmp)
    return liste_tmp

doctest.run_docstring_examples(tri_stupide, globals())   
```

## 2.2. Tri à bulles 


**Principe** : ce tri consiste à parcourir la liste en permutant tous les éléments consécutifs qui ne sont pas dans l'ordre, puis à recommencer tant que la liste n'est pas triée.



<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

**2.2.2.** Chercher sur le Web la complexité en moyenne de cet algorithme.

```python
def tri_bulles(liste):
    """renvoie une copie de la liste triée avec l'algorithme de tri à bulles
    >>> tri_bulles([3,2,1,654,4,2,3,5,8])
    [1, 2, 2, 3, 3, 4, 5, 8, 654]
    >>> tri_bulles([1,2,3,4])
    [1, 2, 3, 4]
    >>> tri_bulles([2,1,3,4])
    [1, 2, 3, 4]
    >>> tri_bulles([1,2,4,3])
    [1, 2, 3, 4]"""
    est_triée = False
    while not est_triée:
        est_triée = True
        for i in range(len(liste)-1):
            if liste[i] > liste[i+1]:
                liste[i], liste[i+1] = liste[i+1], liste[i]
                est_triée = False
    return liste

doctest.run_docstring_examples(tri_bulles, globals())     
% timeit tri_bulles([3,2,1,654,4,2,3,5,8,45,12,3,6,4,568,7,8,98,798,9,46,21,654,96,654])
```

## 2.3. Tri par sélection

**Principe** : ce tri consiste à sélectionner dans la partie de la liste non encore triée le plus petit élément et à le placer en fin de la partie déjà triée.



<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

Déterminer les complexités en temps dans le meilleur cas, en moyenne et dans le pire des cas en justifant la réponse.

*L'algorithme est rappelé ci-dessous*
</div>

```python
def tri_sélection(liste):
    """renvoie une copie de la liste triée avec l'algorithme de tri par sélection
    >>> tri_sélection([3,2,1,654,4,2,3,5,8])
    [1, 2, 2, 3, 3, 4, 5, 8, 654]
    >>> tri_sélection([1,2,3,4])
    [1, 2, 3, 4]
    >>> tri_sélection([2,1,3,4])
    [1, 2, 3, 4]
    >>> tri_sélection([1,2,4,3])
    [1, 2, 3, 4]"""
    liste_tmp = []
    while len(liste) != 0:
        ind_min = 0
        for ind in range(len(liste)):
            if liste[ind] < liste[ind_min]:
                ind_min = ind
        liste_tmp.append(liste[ind_min])
        liste.pop(ind_min)
    return liste_tmp

doctest.run_docstring_examples(tri_sélection, globals()) 
%timeit tri_sélection([3,2,1,654,4,2,3,5,8,45,12,3,6,4,568,7,8,98,798,9,46,21,654,96,654])
```

# 3. Complexité des algorithmes de recherche

## 3.2. Recherche naïve

**Principe** : On parcourt la liste pour comparer chaque élément à l'élément recherché

<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

```python
def rech_naive(élément, liste):
    """Renvoie None si 'élément' n'est pas 'liste', autrement, renvoie son indice
    >>> rech_naive(3,[7,6,5,8,2,3,1,9,4])
    5
    >>> rech_naive(0,[7,6,5,8,2,3,1,9,4])
    """
    for i in range(len(liste)):
        if élément == liste[i]:
            return i
    return None 
    
doctest.run_docstring_examples(rech_naive, globals())   
```

## 3.3. Recherche par dichotomie

**Principe** : La liste étant triée, on regarde si l'élément cherché est plus petit que l'élément central de la liste.

- si c'est le cas on reproduit l'opération avec la première moitié de la liste.
- sinon on reproduit l'opération avec la seconde moitié de la liste

Ainsi de suite jusqu'à trouver ou non l'élément.

<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

**3.3.1.** Déterminer les complexités en temps dans le meilleur cas en justifant la réponse.

*L'algorithme est rappelé ci-dessous*
</div>

```python
def rech_dicho(élément, liste_triée):
    """Renvoie None si 'élément' n'est pas 'liste_triée', autrement, renvoie son indice
    >>> rech_dicho(1,[1,2,4,5,6,7,8,9])
    0
    >>> rech_dicho(9,[1,2,4,5,6,7,8,9])
    7
    >>> rech_dicho(7,[1,2,4,5,6,7,8,9])
    5
    >>> rech_dicho(3,[1,2,4,5,6,7,8,9])
    """
    deb = 0
    fin = len(liste_triée) - 1
    milieu = (deb + fin) // 2
    while deb <= fin :
        if liste_triée[milieu] == élément :
            return milieu
        elif élément < liste_triée[milieu]:
            fin = milieu - 1
        else :
            deb = milieu + 1
        milieu = (deb + fin) // 2
    return None
    
doctest.run_docstring_examples(rech_dicho, globals())  
```

## 3.4. Comparaison des deux algorithmes de recherche

<div style="font-size:1.2em;border-left:10px solid #aaa;border-bottom:10px solid #aaa;padding:10px;">

Exécuter le script suivant pour comparer les deux algorithmes.

Comfirmer avec une analyse des graphiques les résultats précédents.
</div>

```python
def chronometrer():
    from time import time
    from random import randint
    temps_naif, temps_dicho = [], []
    X = []
    for i in range(100):
        n = 100 * i
        liste = [randint(0,n) for k in range(n)]
        liste.sort() #la liste doit être péalablement triée
        #chrono de rech_naive
        debut = time()
        rech_naive(0, liste)
        temps_naif.append(time()-debut)
        #chrono de rech_dicho
        debut = time()
        rech_dicho(0, liste)
        temps_dicho.append(time()-debut)
        X.append(n)
    return X, temps_naif, temps_dicho

%matplotlib inline
import matplotlib.pyplot as plt
plt.figure(figsize=(15,7))
plt.grid(True)

X, temps_naif, temps_dicho = chronometrer()

plt.subplot(121)
plt.plot(X, temps_naif,'-')
plt.title('recherche naïve')
plt.ylabel('temps')
plt.xlabel('taille de la liste')

plt.subplot(122)
plt.plot(X, temps_dicho,'-')
plt.title('recherche par dichotomie')
plt.ylabel('temps')
plt.xlabel('taille de la liste')

plt.show()
```
